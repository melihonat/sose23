### Aufgabe 1: Vorbereitung

```python
import numpy as np
import matplotlib.pyplot as plt
```

Im Folgenden laden wir einen klassischen, aber veränderten Datensatz.

Die Daten findet ihr in `data`. Die erste Dimension (Zeilen) enthält alle Samples und die zweite (Spalten) die vier Merkmale. Die Klasse für jedes der 105 Samples findet ihr in `target`.

```python
dataset = np.loadtxt("Dataset.csv", delimiter=',', skiprows=1)
np.random.shuffle(dataset)

data = dataset[:, :-1]
target = dataset[:, -1]
print(data.shape, np.unique(target))
```

**1.** Vereinfacht das 3-Klassenproblem zu einem 2-Klassenproblem, indem ihr die 3. Klasse für dieses Arbeitsblatt ignoriert.

_Tipp_: Ähnlich wie in Blatt1 können slices und boolean indexes sehr hilfreich sein. Slicing und Indexing Beispiele: https://numpy.org/doc/stable/user/basics.indexing.html#advanced-indexing

```python
two_class_data = []
two_class_target = []

# YOUR CODE HERE
ignore_class2 = target != 2

two_class_data = data[ignore_class2]
two_class_target = target[ignore_class2]

two_class_data.shape, np.unique(two_class_target)
```

**2.** Vervollständigt für diese Aufgabe die Funktion `plot`. Plottet damit die ersten beiden Dimensionen als Punktdiagramm (scatter plot); plottet die ersten beiden Dimensionen anschließend noch einmal, diesmal mit unterschiedlichen Farben für die beiden Klassen.

_Tipp_: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
plt.rcParams['figure.figsize'] = [8, 5]

def plot(x, y, classes=None):
    plt.scatter(x, y)

    if classes is not None:
        unique_classes = np.unique(classes)
        colors = ['blue', 'purple', 'green']

        for i in unique_classes:
            indices = np.where(classes == i)
            plt.scatter(x[indices], y[indices], color=colors[int(i)])

    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.title('Scatter Plot')
    plt.legend()
    plt.show()

# Plotten der ersten beiden Dimensionen
# YOUR CODE HERE
plot(two_class_data[:, 0], two_class_data[:, 1])

# Plotten der ersten beiden Dimensionen mit Klassen in unterschiedlichen Farben
# YOUR CODE HERE
plot(two_class_data[:, 0], two_class_data[:, 1], two_class_target)
```

**3.** Was fällt euch in eurem Plot auf? Sind die Klassen gut trennbar?


Ja, es gibt eine klare imaginäre Linie zwischen den beiden Klassen. Dies wird aber nur durch die Farben deutlich.

### Aufgabe 2: Principal Component Analysis

Wir möchten nun die PCA auf unsere Daten anwenden und verwenden dafür die PCA aus Scikit-learn: 
https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

**1.** Die PCA transformiert Daten, wie tut sie dies?


Sie transformiert die Daten, indem sie neue Variablen, sogenannte principal components, generiert, die eine maximal mögliche Varianz in den Daten erfassen. Die PCA versucht, die ursprünglichen Merkmale in eine neue Merkmalsbasis zu projizieren, die eine geringere Anzahl von Dimensionen (Features) hat.

**2.** Was erwartet ihr von einer Transformation basierend auf eurem obigen Plot?


Wir erwarten, dass die Klassen nach der PCA besser zu unterscheiden sind, da die wichtigsten Features für die Klassentrennung durch die PCA hervorgehoben werden.

**3.** Transformiert die Daten mithilfe der PCA und plottet wieder die ersten beiden Dimensionen.

```python
from sklearn.decomposition import PCA

pca = PCA()

# YOUR CODE HERE
two_class_data = pca.fit_transform(two_class_data)

plot(two_class_data[:, 0], two_class_data[:, 1], two_class_target)
```

**4.** Entspricht diese Transformation euren Erwartungen? Warum/warum nicht? 


Ja, da es jetzt nicht nur eine klare imaginäre Linie zwischen den beiden Klassen gibt, sondern auch sehr viel Platz zwischen ihnen.

**5.** Wie viel Varianz wird durch die erste transformierte Dimension erklärt?

```python
explained_var = 0

# YOUR CODE HERE
explained_var = pca.explained_variance_ratio_

print(explained_var[0])
```

```python
# Euer Ergebnis wird automatisch getestet.
```

_Optional_: Plottet die Varianz pro Dimension, was fällt euch auf?

```python
# YOUR CODE HERE
plt.bar(range(len(explained_var)), explained_var)
plt.xlabel('Transformed Dimension')
plt.ylabel('Variance explained')
plt.show()

# Es fällt auf, dass nach der ersten transformierten Dimension nicht mehr viel passiert.
```

### Aufgabe 3: Linear Discriminant Analysis

Wir möchten nun eine LDA auf unsere Daten anwenden und verwenden dafür die LDA aus Scikit-learn: 
https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html

**1.** Wie unterscheidet sich die LDA von der PCA? Wie transformiert sie den Merkmalsraum?


PCA ist ein unüberwachtes Verfahren, das auf der Varianz der Daten basiert und keine Informationen über die Klassenetiketten verwendet.
Die LDA jedoch ist ein überwachtes Verfahren, das die Klassenetiketten der Daten nutzt, um die Klassenunterscheidung zu maximieren.

Während die PCA die Daten in einen neuen Merkmalsraum transformiert, der aus den Hauptkomponenten besteht, transformiert die LDA die Daten in einen neuen Merkmalsraum, der so gewählt wird, dass die Klassenunterscheidung maximiert wird.

**2.** Basierend auf eurem farbigen Plot oder der PCA, wie viele Dimensionen benötigt ihr vorraussichtlich um die Klassen vernünftig zu trennen?


Um nur eine Klassentrennung zu gewährleisten, würde theoretisch sogar eine einzige Dimension genügen, da genug Platz zwischen den Elementen der beiden Klassen ist.

**3.** Führt nun eine LDA durch. Überschreibt hierzu die Anzahl der Komponenten basierend auf eurer Antwort oben. Transformiert anschließend die Daten und plottet das Ergebnis wieder farbig.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=1)

# YOUR CODE HERE
two_class_data_lda = lda.fit_transform(two_class_data, two_class_target)

plt.hist(two_class_data_lda[two_class_target == 0.0])
plt.hist(two_class_data_lda[two_class_target == 1.0])
```

**4.** Entspricht diese Transformation euren Erwartungen? Warum/warum nicht?

Ja, weil die Distanz zwischen den Elementen eindeutig auf eine Trennung hindeutet.

**5.** Die LDA kann auch zur Klassifikation, also der automatischen Zuordnung eines Samples zu einer Klasse, genutzt werden. Überlegt euch ein einfaches Konzept (basierend auf eurem Plot oben) diese beiden Klassen zu trennen.

_Hinweis_: Das Konzept muss nicht programmiert werden. Eine schriftliche Antwort reicht.


Eine Möglichkeit für ein Konzept zur Klassifikation eines Samples, ist, den Wert des Samples auszulesen und zu schauen ob er positiv oder negativ ist. Demnach können dann die Samples mit positiven Werten der Klasse in orange zugewiesen werden und umgekehrt.

**6.** Gibt es je eine Möglichkeit, dass
- eine PCA gut auf einem Datensatz implementiert werden kann, aber eine LDA nicht?
- eine LDA auf einem Datensatz gut implementiert werden kann, aber eine PCA nicht?

Wenn ja, nennt ein Beispiel. Wenn nein begründet, weshalb nicht.


PCA gut, LDA nicht: Wenn die Klassenlabels in den Daten keine klare Trennung zwischen den Klassen aufweisen und die Hauptvarianz in den Daten nicht stark von den Klassen abhängt, kann die PCA dennoch sinnvoll angewendet werden, um die Dimensionalität zu reduzieren und Muster in den Daten zu identifizieren. Die PCA zielt darauf ab, die Varianz insgesamt zu maximieren, unabhängig von den Klassenlabels.

LDA gut, PCA nicht: Wenn die Klassenlabels in den Daten eine starke Trennung zwischen den Klassen aufweisen und das Hauptziel darin besteht, eine optimale Trennung zwischen den Klassen zu erreichen, kann die LDA bessere Ergebnisse liefern als die PCA. Die LDA nutzt die Klasseninformationen, um den Merkmalsraum so zu transformieren, dass die Klassen besser separiert werden können.

Ein Beispiel für den ersten Fall könnte ein Datensatz sein, in dem die Klassenlabels zwar vorhanden sind, aber keine eindeutige Trennung zwischen den Klassen besteht, zum Beispiel bei einer Aufgabe der Gesichtserkennung, bei der die Merkmale (z.B. Pixelintensitäten) stark überlappen.

Ein Beispiel für den zweiten Fall könnte ein Datensatz sein, bei dem die Klassenlabels stark unterschiedliche Gruppen repräsentieren, z.B. bei einer medizinischen Diagnoseaufgabe, bei der die Klassen "gesund" und "krank" deutlich unterschiedliche Merkmale aufweisen.
